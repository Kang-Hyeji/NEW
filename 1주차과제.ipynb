#음성

인공지능 제품이나 서비스를 떠올릴 때, 너무 당연하다고 인식되는 기능이 바로 '음성인식'이다. 음성인식 기술은 최근 빠른 발전을 이뤄냈고 우리에게 편리함을 제공해주고 있다. 스마트폰의 음성 기반 인공지능 비서는 이름만 불러도 기다렸다는 듯 인사를 건넨다. 차 안에 한 대씩은 있는 자동차 내비게이션은 물론 인공지능 스피커도 최근에는 흔히 볼 수 있게 되었다. 덕분에 말 한마디를 건네서 날씨나 지리 정보를 쉽게 얻을 수 있고 기계와 농담까지 주고받을 수 있게 되었다. 이렇게 음성인식 제품들은 기술과 사람을 더 가깝고 친근하게 만들어 준다. 사람에겐 너무도 간단한 대화를 음성인식 제품들과 나눌 때, 그 안에서는 꽤 복잡한 일들이 벌어지고 있다.
우리가 인공지능 기반 음성인식 제품에 말을 하면, "음성 입력 및 인식→자연어 처리→인식결과"의 단계를 거친다. 기계가 사람의 언어를 인식하고 이해하는 데 필요한 과정이다. 첫 번째 과정에서는 음성 입력 과정을 거쳐 사람의 음성을 컴퓨터가 이해할 수 있도록 텍스트화한다. 이 기술을 STT(Speech To Text)라고 하는데, 어린아이가 받아쓰기하듯이 사람의 음성을 텍스트로 옮기는 것이다. 주변 소음을 제외한 말소리를 파악하고 이어서 각 발음과 단어를 인식한 후 핵심어와 연결 단어를 인식해 입력해야 하므로 생각보다 간단하고 쉽지만은 않다.
두 번째는 자연어 처리(Natural Language Processing, NLP)과정이다. 자연어란 사람이 의사소통을 위해 사용하는 언어이다. 자연어 처리란 컴퓨터가 자연어를 분석하여 이해하고 처리하는 기술이다. 이 과정에서는 자연어에 대한 형태소 분석, 구문 해석, 의미 분석, 화용 분석 등을 통해 컴퓨터가 문장에 담긴 의도를 파악하게 한다. 이 자연어 처리는 인공지능의 주요 분야이다. 인공지능 기계가 사람의 언어를 얼마나 잘 파악하였는지를 알 수 있는 과정이다.
마지막으로 기계는 인식 결과를 내놓는다. 인식된 요청에 따라 가장 최적의 결과를 찾아내고, TTS(Text To Speech) 기술을 통해 사람의 말소리처럼 응답하는 것이다. 친구에게 메시지를 보내라고 말했다면 전달한 내용으로 메시지를 전송하고, 날씨를 물어봤다면 인터넷으로 위치 기반의 날씨를 검색해 대답해준다. 음성인식 기술은 인공지능 딥러닝 기술이 뒷받침되어 최근 크게 발전하였다. 많은 양의 데이터를 컴퓨터에 입력시키고 학습시킨 뒤 자연어 처리를 통해 사람의 언어를 제대로 이해하고 반응하게끔 한다.
인공지능의 자연어 처리 기술의 진화에 따라 더 똑똑한 음성인식 서비스가 탄생할 수 있다. 많은 발전을 이룬 음성인식 기술이지만, 아직도 한계는 존재한다. 사람마다 발음, 억양, 톤이 다르고, 또 일상 대화에서는 문법에 어긋난 문장도 많고 주변의 소음도 섞이는데 현재는 기술이 이 모든 것을 구분해 인식하지 못하기 때문에 음성인식의 정확도에 대한 이슈가 있는 상황이다. 이러한 한계를 극복하기 위해서는 더 많은 데이터를 확보해야 한다. 주변 소음이 담겨있는 음성이나 우리가 일생생활에서 쓰는 구어체 음성, 의미는 통하지만, 문법에 맞지 않는 문장에 대한 음성 등 다양한 음성 언어 데이터를 확보해야 인공지능의 음성인식 기술 성능을 향상시킬 수 있다.

음성합성 기술(TTS:Text-To-Speech)은 지하철이나 버스 정류장의 안내방송, 아파트관리 사무소, ATM기에서 나오는 목소리, 은행에 전화를 걸면 흘러나오는 ARS 음성 등이 모두 음성합성 기술로 만들어낸 목소리이다. 최근 이런 음성 합성 기술에 인공지능을 결합해 실제 화자와 구별하기 힘들 정도의 자연스럽고 매끄러운 AI보이스를 만들 수 있게 되었다. 애플의 실나 구글 어시스턴트의 경우도 이런 음성합성 기술을 활용한 대화 인공지능이다. 이런 TTS 기술은 다양한 분야에서 널리 이용되고 있으나 특히 가장 두드러지게 활용되고 있는 분야는 바로 엔터테인먼트 업계이다. 성우의 오랜 녹음 시간이 있어야 하는 오디오북, 발 빠르게 최신 유행을 전달해야 하는 1인 방송인, 인건비 문제를 겪는 라디오 새벽 방송 또는 재난 방송과 같은 채널 등이 있다. 
음성합성 기술이란, 문자 정보를 음성 정보로 바꿔주는, 쉽게 말해 컴퓨터가 음성을 만드는 기술을 일컫는다. 텍스트를 음성으로 변환해, 인위적으로 사람의 음성을 합성, 실시간 커뮤니케이션 기술을 현실화한 대화형 인공지능 서비스이다. 실제로 maum.ai의 TTS 엔진은 딥러닝을 통해 음성 녹음 시간을 단축하되, 실제 그 사람의 목소리 그대로 자연스럽게, 세계 최고 수준의 음질과 실시간 합성 속도를 제공하고 있다.



#이미지

이미지 분석 기술은 기존에도 컴퓨터비전을 중심으로 연구되어 온 분야였다. 이것이 최근에는 더욱 발전된 인공신경망 기술인 딥러닝을 통해 사람의 시각처럼 높은 정확도와 다양한 사물 인식이 가능한 AI 이미지 분석으로 한 단계 진화하여 등장하였다. 인물이나 피사체의 움직임을 정교하게 인식하여 촬영하는 카메라,사진을 '음식','동물'등으로 자동 분류해 주는 구글 포토, 카메라를 비추면 어떤 사물인지 알려주는 구글 렌즈 등이 실생활에서 찾아볼 수 있는 AI 이미지 분석 기술의 사례이다. AI이미지 분석의 핵심 기술은 바로 '노이즈(Noise)'를 얼마나 판별해 내느냐 하는 것이다. 예를 들어,판다(동물) 사진에 노이즈를 삽입했을 경우, 사람은 여전히 판다로 정확히 판단하지만, AI는 각각의 데이터(픽셀)로 이미지를 분석하기 때문에 긴팔원숭이로 잘못 인식하는 경우가 있다. 또한, 노이즈는 아니지만 객체의 구도와 모양, 그리고 색이 유사하여 이미지 분석을 혼란스럽게 만드는 경우들도 있다. 노이즈 외 판별에 핵심이 되는 요소로는, 판별하고자 하는 이미지의 특징을 포착하고 추상화하는 것이다. 예를 들면 고양이 얼굴의 경우 뾰족한 귀, 가는 눈, 수염 등 고양이를 대표하는 특징을 잘 포착하고 이를 추상화하여 다양한 품종의 고양이 얼굴을 모두 고양이로 판별하게끔 만들어야 한다.
 인공지능을 통한 이미지 분석이 각광받고 있는 것은 다양한 산업 분야에 활용할 수 있기 때문이다. 생산시설, 무인 스토어, 보안 검색대까지 AI 이미지 분석은 유용함을 넘어 이제 필수로 여겨지고 있다.
 
 '인공지능 숫자 인식 시스템'만 하여도 많은 분야에 적용할 수 있으며, 이중에서 대표적인 분야가 배송업체다. 고객이 물품에 적어놓은 우편번호나 전화번호 등의 기록을 사진으로 촬영하고 정보를 읽어 들여 고객정보와 수신처를 자동으로 전산 등록하며 지역을 분류해 낼 수 있다. 과거에는 물건이 거점센터에 집결되면 직원이 직접 컴퓨터에 정보들을 입력하고 데이터 라벨을 인쇄하여 부착하는 방식이었다. 하지만, 이것이 운송업체에 적용되기 위해서는 훨씬 더 높은 수준의 정확성이 요구된다. 흐릿하거나 기울어져 촬영된 사진, 입력 칸에 걸쳐 있거나 벗어나 쓰인 숫자, 사인펜으로 두껍게 쓰여 뭉쳐 보이는 숫자 등 다양한 글씨체와 상황에 대응해서도 99.99%이상의 신뢰성을 요구하고 있다. 이러한 문제점들은 현제 수학적인 이미지 분석기법을 통하여 거의 해결된 상태다.
현재까지 상용화된 제품들의 기술 수준으로 미루어볼 때, 필기체 문장 인식은 조금 더 개발의 시간이 필요해 보인다. 일단, 인쇄체에 대한 영문의 인식률은 상당히 높고, 한글 역시 우수한 편이다. 하지만 필기체에 대한 인식은 다양한 글씨체를 형태학적으로 분류하기가 쉽지않다. 특히, 영문보다 한글이 더욱 어렵기 때문에 한글 필기체 자동 인식은 좀 더 시일이 걸릴 것으로 예상된다.
 AI 이미지 인식 기술은 사람보다 더 정확해지고 있지만, 문제는 AI가 그만큼 많은 양의 데이터를 학습해야만 한다는 점이다. 사람은 몇 장의 사진만으로도 어떤 사물인지 쉽게 학습할 수 있는데, 현재의 AI는 수만 장의 사진을 학습해야 하기 때문이다. 더욱이 AI가 내놓는 결과의 타당성을 최종적으로는 사람이 판단해야한다. 이런 부분에서 아직은 사람이 AI의 선생님 역할을 해야 하는 상황이라고 할 수 있다. 이러한 부분들이 아직 극복된 건 아니지만, 최근에는 마치 사람처럼 ‘학습하는 방법 자체를 학습하는 기술(메타 학습, Meta Learning)’을 도입하여 문제를 해결하려는 긍정적인 시도들이 있다.

#자율주행

현재 자율주행은 실리콘밸리에서 가장 트렌디한 기술 중 하나이다. 인공지능의 한 분야인 머신러닝이 발전하면서 현실 세계에서 수집한 방대한 데이터를 흡수한 자동차는 스스로 운전을 배울 수 있게 되었다. 자동차들은 자율주행을 하면 할수록 더 많은 양의 데이터를 수집해 주행 성능을 개선한다. 스마트폰 애플리케이션 터치 한번으로 부를 수 있는 로보택시(Rovotaxis, 자율주행 자동차에 택시 서비스가 결합한 것)는 자동차의 소유권이라는 개념을 사라지게 할 것이다. 무엇보다 전자 기기의 속도로 작동하는 자율주행차의 반사 신경은 안전성을 획기적으로 끌어올릴 수 있을 것으로 기대된다. 자동차와 기술 업계 리더들은 '무사고'가 실현되는 세상을 꿈꾸고 있다.
자율주행 기술의 실현은 이미 우리 눈앞에 와있지만, 여전히 불확실한 상태이다. 2015년 테슬라의 최고경영자인 일론 머스크는 2017년까지 테슬라의 자동차들이 "완전히 자율적으로" 움직일 수 있을 것이라고 예견했다. 2018년 제너럴모터스는 2019년에 핸들이나 페달을 장착하지 않은 자동차를 선보일것이라고 발표했다. 이는 실현되지 못했다. 알파벳의 자회사이자 자율주행 업계 선두주자인 웨이모(Waymo)는 자율주행 자동차를 시범 운영하고 있었던 미국 피닉스에서 2018년 말까지 기사 없는 택시를 출시하기 위해 총력을 기울였지만 계획은 실패로 끝났다.
현재 가장 주목받는 기술인 자율주행 산업의 발전이 정체를 겪는 이유는 무엇일까? 문제를 해결하기 위해 중국은 어떤 노력을 기울이고 있을까? 인간 운전자의 역량에 버금가는 소프트웨어를 만들어 내는 일은 여전히 어렵다. 인공지능 기술은 혁신적이지만, 동시에 제한적이다. 기대만큼 빠른 속도로 기술이 개발되지 않으면서, 중국은 새로운 대안을 마련하고 있다. 바로 도로와 도시를 뜯어고치는 것이다. 낮은 수준의 소프트웨어로도 주행할 수 있는 단순한 도로 시스템, 사고를 방지하기 위한 보행자 안전 규칙의 신설, 자율주행기업의 사고에 대한 법적 책임 경감 등이 핵심이다. 
과거 단순히 이동수단으로 치부되던 자동차가 미래 모빌리티 플랫폼이라는 개념이 생기면서 '움직이는 생활공간', '제2의 거실'이라 칭해지고 있다. 미래의 최고 화두는 전기차와 수소연료전지차, 자율주행 자동차, 그리고 이를 섞은 공유경제인 모빌리티 쉐어링일 것이다.이 중에서도 자동차를 포함해 미래 모빌리티를 제어하는 자율주행 기능이 가장 화제가 되는 것은 인공지능을 적용하거나 공유경제 모델을 극대화하는 등 다양한 활용도를 기대할 수 있기 때문이다. 최근 자율주행차는 괄목할만한 성장을 거듭하고 있다. 자율주행은 레벨 0부터 레벨 5까지 6단계로 나뉘며, 우리가 생각하는 진정한 자율주행 기능은 레벨 4와 레벨 5 이다. 이 상태가 되면 사고 없는 완벽한 이동시대로 접어들 수 있으며, 혹시 교통사고가 발생해도 자동차에 책임을 물을 수도 있기 때문이다. 현재 적용되는 자율주행 기술은 레벨 2정도이며 이 단계는 한산한 고속도로라면 잠시 운전대에서 손을 떼고 병을 따거나 옆 좌석의 물건을 집는 등 간단한 행동을 해도 안전하다. 이른바 운전자 보조운행 시스템인 ADAS라 할 수 있다. 올해 후반에는 고급차를 중심으로 레벨 3가 적용된다고 한다. 이 레벨은 일정 지역에서 날씨가 좋은 경우 운전대에서 손을 놓고 안전하게 간단한 일을 할 수 있다고 한다. 
자율주행 기술개발에서 가장 핵심 과제는 판단기능에 인공지능을 포함하는 것이다. 사람과 같이 학습기능을 가진 인공지능을 포함할 경우 더욱 완벽한 운전이 가능해진다. 물론 해결 과제도 아직 많다. 학습기능에 따라 사람의 감성적인 기능이나 윤리적인 판단과는 전혀 다른 방향으로 자동차가 움직일 수 있기 때문이다. 예를 들면 횡단보도 위를 차례대로 손을 들고 걷고 있는 4명의 어린이를 마주한 운전자가 시속 100km에서 브레이크가 고장 났을 경우, 사람이 운전했을 때 탑승자 3명을 고려해도 무조건 핸들을 꺾어서 아이들을 피할 것입니다. 그러나 인공지능이 포함된 자율주행차는 아이 중 가장 간격이 큰 아이를 대상으로 한 명만 치고 지나가면서 탑승자 3명을 구했다고 판단할 수도 있습니다. 이러한 경우 윤리적으로 심각한 논란이 예상되는데요. 이미 비슷한 사례가 무수히 많은 만큼 인공지능이 포함된 자율주행차가 어떻게 운행될 것인지에 대해 고민이 필요합니다. 
자율주행 기술이 개발되고 있지만, 아직까지 운전은 분업이 불가능하기도 하다. 처음에는 자율주행 기능을 신뢰하지 않아 운전대를 잡고 브레이크를 밟을 준비를 하고 있을 것이다. 하지만 자율주행 기능에 익숙해진 운전자는 점차 안심하고 휴대폰 보는 시간이 길어지게 된다. 그러다 결국은 자율주행에 완전히 운전을 맡기고 영화를 보는 상황이 발생하게 될 것이다. 이미 국내에서도 유사 사고가 여러 건 발생했고, 본격적으로 레벨 3의 기술이 적용되면 이러한 사례는 더욱 늘어날 것으로 예측된다. 특히 운전자가 기술의 완벽한 부분만을 보고 접하는 상황에서, 자동차 판매현장도 자사 차량의 자율주행 기능만을 강조하고 있어서 더욱 우려가 된다.
현재 자율주행자동차는 약 200m 주위를 레이다(Radar)센서, 초음파 센서, 라이다(Lidar)센서를 비롯해 여러 카메라 등 차량 주변 수십 개의 센서를 이용해 주변 정보를 인지하고 있다. 그러나 아직 시스템적으로 완전한 구현은 어려운 상황으로, 주변 도로 환경의 인지에 한계를 드러낸다. 먼지가 많은 오프로드나 폭우, 폭설은 물론, 도로 위의 포트홀 등 상황에 따라 오류나 인지 불가능 등이 발생할 수 있다. 따라서 날씨가 맑은 낮 시간대 한산한 도로 위의 운전으로만 한정해 사용이 가능하다. 자율주행차운행을 위한 법적·제도적 부분도 아직은 미흡하다. 자율주행차가 본격 등장하면 차량이 법적 인격체가 되면서 사고 책임과 보험 청구 등에 대해 새로운 규제가 생겨나게 된다. 차량의 완성도와 안전성 등이 자율주행의 핵심이지만, 인공지능과 자율주행에 대한 법적·제도적 기반 또한 갖춰줘야 할 것이다.
인공지능을 기반으로 한 자율주행차는 단순히 차량만 해당하는 것이 아니라 플라잉카 등 모든 모빌리티개념으로 확대할 수 있다. 특히 자율주행 기능은 자율주행 전용 고속도로 등을 이용한 군집운행이나 아예 인간 운전과는 구분되는 공중운행 개념의 플라잉 카에 적용시키기 용이하다. 예전에는 2020년 하면 하늘을 날고 특별한 운송장치가 활약하는 상상을 하는 경우가 많았다. 현재 그러한 단계까지 발전되지는 못했지만 인공지능과 자율주행의 발전 속도를 볼 때, 앞으로 10년이 지난 2030년에는 우리가 상상하던 하늘을 나는 자동차가 실제로 구현될 수도 있을지 모른다.
